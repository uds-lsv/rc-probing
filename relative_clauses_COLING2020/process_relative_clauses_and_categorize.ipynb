```python
# read WHO and WHICH word list

with open('WHO.words') as f:
     WHO_words = f.readlines()
        
with open('WHICH.words') as f:
     WHICH_words = f.readlines()
```

```python
RELATIVIZERS = {'who', 'which', 'whom', 'that'}

HUMAN_RELS = {'who', 'whom'}
```

```python
df_xx = list()

for idx, row in df_X.iterrows():
    
    # ignore RCs that do not start with a relativizer

    
    if row['RC'].split(' ')[0] not in RELATIVIZERS:
        continue
        
    ### identify HUMAN vs. NON-HUMAN relativizer
    if row['relativizer_word'] in HUMAN_RELS:
        row['HUMAN'] = 1
        
    elif row['relativizer_word'] == 'which':
        row['HUMAN'] = 0
        
    elif row['relativizer_word'] == 'that':
        # check for a WHO word
        if row['RC_modifier_head'] in WHO_words:
            row['HUMAN'] = 1            
        
        # check for a WHICH word
        elif row['RC_modifier_head'] in WHICH_words:
            row['HUMAN'] = 0
        
        # else assign 'NaN'
        else:
            row['HUMAN'] = 'NaN'  
            continue 
            
            
    ### identify RESTRICTIVE vs. NON-RESTRICTIVE relativizer
    # if left context word is comma, the it is NON-RESTRICTIVE
    if make_tuple(row['RC_context'])[0] in {",", "-"}:        
        row['RESTRICTIVE'] = 0
        
    elif row['relativizer_word'] == 'that':
        row['RESTRICTIVE'] = 1  
        
    # these cases are NOT 100% guaranteed to be RESTRICTIVE, but we tried our best
    elif row['relativizer_word'] in {'who', 'which', 'whome'}:
        row['RESTRICTIVE'] = 1
        
    # else assign 'NaN'
    else:
        row['RESTRICTIVE'] = 'NaN'  
        continue
             
    # append to dataframe 

    
    df_xx.append(row)

    
    #print(idx, df_xx[-1]['RC'])
    #if idx > 150:
        #break
    
```

```python
MODIFICATIONS = {
    'HRSubj': {'WHO2WHICH': 0, 'NULLREL': 0, 'IDENTITY': 1}, 
    'HRObj': {'WHO2WHICH': 0, 'NULLREL': 1, 'IDENTITY': 1}, 
    'NHRSubj': {'WHICH2WHO': 0, 'NULLREL': 0, 'WHICH2THAT': 1, 'IDENTITY': 1}, 
    'NHRObj': {'WHICH2WHO': 0, 'NULLREL': 1, 'WHICH2THAT': 1, 'IDENTITY': 1}, 
    'HNR': {'WHO2WHICH': 0, 'NULLREL': 0, 'IDENTITY': 1}, 
    'NHNR': {'WHICH2WHO': 0, 'NULLREL': 0, 'IDENTITY': 1}
}
```

```python
MODIFICATIONS['HRSubj']
```

```python
def modify_sentence(instance):
    """Given a sentence, modify according to category 
        then return modified plus label. """
    
    human = instance['HUMAN']
    restrictive = instance['RESTRICTIVE']
    is_subj_RC = instance['relativizer_func'] in ['nsubj', 'nsubjpass']
    is_obj_RC =  instance['relativizer_func'] in ['pobj', 'dobj']
    
    if human == 0 and restrictive == 0:
        category = 'NHNR'
        
    elif human == 0 and  restrictive == 1:
        if is_subj_RC: category = 'NHRSubj'
        elif is_obj_RC: category = 'NHRObj'
        else: return None
        
    elif human == 1 and  restrictive == 0:
        category = 'HNR'
        
    elif human == 1 and  restrictive == 1:
        if is_subj_RC: category = 'HRSubj'
        elif is_obj_RC: category = 'HRObj'
        else: return None
        
    else:  # if there a NaN then just continue  
        return None
         
        
    # sample a legitimate modification
    n = len(MODIFICATIONS[category])
    _rand =  math.floor(n*random.random())   
    _modify = list(MODIFICATIONS[category].keys())[_rand]
    
    
    print(_modify)
    
    sentence = instance['sentence'].split()
    
    # check if WHO2WHAT or friends cases
    if _modify.count('2') > 0:
        
        actual_rel = instance['relativizer_word'] 
        
        # get the corrputor relativizer
        corrupt_rel = _modify.lower().split('2')[1]
        
        actual_idx = sentence.index(actual_rel)
        modified = sentence[:actual_idx] 
        modified += [corrupt_rel]
        modified += sentence[actual_idx + 1:]
        
        modify_sentence = ' '.join(modified)
        
    
    elif _modify == 'NULLREL':
        actual_rel  = instance['relativizer_word']      
        actual_idx = sentence.index(actual_rel)
        modified = sentence[:actual_idx] 
        modified += sentence[actual_idx + 1:]
        modify_sentence = ' '.join(modified)
        
        
    elif _modify == 'COMMAINSERT':
        RC  = instance['RC'].split()
        
        actual_rel  = instance['relativizer_word']      
        actual_idx  = sentence.index(actual_rel)        
        
        modify_RC = [', '] + RC
        
        if sentence[actual_idx + len(RC)] not in string.punctuation: 
            modify_RC = modify_RC + [', ']
        
        modified = sentence[:actual_idx] + modify_RC
        modified += sentence[actual_idx + len(RC):]
        modify_sentence = ' '.join(modified)
        
    elif _modify == 'IDENTITY':
        modify_sentence = ' '.join(sentence)
    
    return MODIFICATIONS[category][_modify], _modify, modify_sentence
```

```python
df_list = list()

for idx, row in df_xx.iterrows():
    
    #print(row)
    try:
    
        # original sentence 
        sentence = row['sentence']
        #label, modification, modified_sentence = (1, 'ORIGINAL', sentence) 
        #df_list.append((label, row['HUMAN'], row['RESTRICTIVE'], modification, sentence, modified_sentence))
        # modified sentence

        if modify_sentence(row):
            label, modification, modified_sentence = modify_sentence(row)

            df_list.append(
                (
                    label, 
                    row['HUMAN'],
                    row['RESTRICTIVE'],
                    modification,
                    sentence, 
                    modified_sentence,
                    row['RC'],
                    row['RC_subj'],
                    row['RC_subj_pos'],
                    row['relativizer_word'],
                    row['relativizer_func'],
                    row['RC_modifier_head'],
                    row['RC_modifier_NP']
                )
            )

    
        
    except:
        pass 
    
```
